```python
def pplot(df,i):
    plt.figure(figsize=(20,5))
    plt.subplot(1,3,1)
    sns.histplot(x=i,data=df,kde=True)
    plt.subplot(1,3,2)
    sns.regplot(x=i,y='W',data=df)
    plt.subplot(1,3,3)
    sns.boxplot(y=i,data=df)
    plt.show()
```

```python
pplot(df,'R')
pplot(df,'AB')
pplot(df,"H")
pplot(df,'2B')
pplot(df,'3B')
pplot(df,'HR')
pplot(df,"BB")
pplot(df,'SO')
pplot(df,'SB')
pplot(df,'RA')
pplot(df,'ER')
pplot(df,'ERA')
pplot(df,'CG')
pplot(df,'SHO')
pplot(df,'SV')
pplot(df,'E')
```

```python
x=df.iloc[:,1:16]
x
```

```python
y=df.iloc[:,0]
y
```

```python
print(x.shape)
print(y.shape)
```

```python
model=[DecisionTreeRegressor(),KNeighborsRegressor(),AdaBoostRegressor(),LinearRegression()]
max_r2_score=0
for r_state in range(40,100):
    train_x,test_x,train_y,test_y=train_test_split(x,y,random_state=r_state,test_size=0.33)
    for i in model:
        i.fit(train_x,train_y)
        pre=i.predict(test_x)
        r2_sc=r2_score(test_y,pre)
        print('R2 score correspond to random state',r_state,'is',r2_sc)
        if r2_sc>max_r2_score:
            max_r2_score=r2_sc
            final_state=r_state
            final_model=i
            
            
print()
print()
print()
print()
print('Max R2 score correspond to random state',final_state,'is',max_r2_score,'and model is ',final_model)
```

Max R2 score correspond to random state 62 is 0.8269895707447674 and model is  LinearRegression()

```python
KNN=KNeighborsRegressor(n_neighbors=2)
KNN.fit(train_x,train_y)
pred=KNN.predict(test_x)
r2_sc=r2_score(test_y,pred)
print('R2 score',r2_sc*100)
```

```python
print('mean absolute error:',mean_absolute_error(test_y,pred))
print('mean squared error:',mean_squared_error(test_y,pred))
print('Root mean abosulte error:',np.sqrt(mean_absolute_error(test_y,pred)))
```
